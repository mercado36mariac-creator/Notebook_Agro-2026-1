### 1. Cálculo del Índice de Madurez

En la industria de jugos o frutas frescas, se usa la relación entre sólidos solubles ($^\circ Brix$) y acidez para determinar el momento óptimo de cosecha.

```
# Datos de la muestra
grados_brix = 12.5
acidez_titulable = 0.8  # porcentaje de ácido cítrico

# Cálculo del índice
indice_madurez = grados_brix / acidez_titulable

print(f"El índice de madurez es: {indice_madurez}")

if indice_madurez > 15:
    print("Estado: Fruta muy dulce, ideal para consumo inmediato.")
else:
    print("Estado: Fruta ácida, requiere más tiempo o proceso industrial.")
```

### 2. Control de Inventario de Granos (Silox)

Un script simple para monitorear la capacidad de almacenamiento y emitir alertas.

```
capacidad_maxima = 5000  # Toneladas
inventario_actual = 4800

def verificar_alerta(cantidad):
    porcentaje = (cantidad / capacidad_maxima) * 100
    if porcentaje >= 90:
        return "¡ALERTA! Silo casi lleno. Detener recepción."
    else:
        return "Capacidad operativa normal."

print(verificar_alerta(inventario_actual))
```

### 3. Estimación de Evapotranspiración (Simplificada)

Calcular cuánta agua necesita un cultivo es fundamental para el riego de precisión.

```
# Estimación rápida de necesidad de riego diaria (mm)
evapotranspiracion_base = 5.2  # mm/día según clima
coeficiente_cultivo = 0.85     # Específico para el tipo de planta (Kc)

agua_necesaria = evapotranspiracion_base * coeficiente_cultivo

print(f"El cultivo necesita {agua_necesaria:.2f} mm de agua hoy.")
```
