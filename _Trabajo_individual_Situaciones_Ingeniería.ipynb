**Capacidad de Silos de Granos**

**Plantamiento del problema:**

Se requiere calcular cuántas toneladas de maíz puede almacenar un silo cilíndrico. El usuario ingresa el radio y la altura del silo en metros.

**Requerimientos del problema:**

1. Ingresar el radio del silo (float).
2. Ingresar la altura del silo (float).
3. Calcular el volumen del cilindro:

$$V=π*r^2*h$$
4. Calcular la capacidad usando la densidad del maíz (0.75 ton/m³).

**Prueba del programa:**

 Entrada:

*   Radio = 3m
*   Altura = 10m.

 Salida ≈212.06 toneladas.


```python
import math

# ENTRADA
r = float(input("Ingrese el radio del silo en metros: "))
h = float(input("Ingrese la altura del silo en metros: "))

# PROCESO
V = math.pi * (r ** 2) * h
capacidad = V * 0.75

# SALIDA
print("La capacidad del silo es de", capacidad, "toneladas")
```

### Explicación detallada del código para el cálculo de la Capacidad de Silos de Granos

**Sección ENTRADA:**
* `import math` = Esta línea importa el módulo `math` de Python. Este módulo proporciona acceso a funciones y constantes matemáticas, como `math.pi`, que representa el valor de Pi (π). Se utiliza para calcular el volumen del cilindro, ya que la fórmula incluye Pi.
*   `r = float(input("Ingrese el radio del silo en metros: "))`: Esta línea solicita al usuario que ingrese el valor del radio del silo en metros. La función `input()` lee la entrada del usuario como una cadena de texto, y `float()` la convierte en un número decimal (flotante) que se almacena en la variable `r`.
*   `h = float(input("Ingrese la altura del silo en metros: "))`: De manera similar a la anterior, esta línea pide al usuario que ingrese la altura del silo en metros. La entrada se convierte a un número flotante y se guarda en la variable `h`.

**Sección PROCESO:**
*   `V = math.pi * (r ** 2) * h`: Aquí se calcula el volumen del silo cilíndrico. `math.pi` proporciona el valor de π (aproximadamente 3.14159), `r ** 2` calcula el radio al cuadrado, y luego se multiplica por la altura `h`. El resultado (el volumen en metros cúbicos) se almacena en la variable `V`.
*   `capacidad = V * 0.75`: Esta línea calcula la capacidad de almacenamiento del silo en toneladas. Se multiplica el volumen `V` (en m³) por la densidad del maíz (0.75 toneladas por metro cúbico). El resultado se guarda en la variable `capacidad`.

**Sección SALIDA:**
*   `print("La capacidad del silo es de", capacidad, "toneladas")`: Finalmente, esta línea muestra el resultado al usuario. Imprime un mensaje descriptivo junto con el valor calculado de la `capacidad` y la unidad de medida "toneladas".

**Dosificación de Fertilizante Líquido**

**Planteamiento del Problema:**

Un sistema de riego por goteo requiere aplicar una dosis de fertilizante por hectárea. El usuario ingresa las hectáreas totales y la dosis requerida (litros/hectárea).


**Requerimientos del problema:**

1. Ingresar hectáreas (float).
2. Ingresar dosis (L/ha).
3. Calcular fertilizante total.
4. Calcular agua usando relación 1:500.

**Prueba del programa:**

Entrada:

*   5 hectareas
*   2 L/ha

Salida esperada:

*   10 L fertilizante
*   5000 L agua



```python
# ENTRADA
hectareas = float(input("Ingrese las hectáreas: "))
dosis = float(input("Ingrese la dosis (L/ha): "))

# PROCESO
fertilizante = hectareas * dosis
agua = fertilizante * 500

# SALIDA
print("Se necesitan", fertilizante, "L de fertilizante")
print("Se necesitan", agua, "L de agua")
```

### Explicación detallada del código para la Dosificación de Fertilizante Líquido

**Sección ENTRADA:**
*   `hectareas = float(input("Ingrese las hectáreas: "))`: Solicita al usuario que ingrese el número total de hectáreas. La entrada se convierte a un número decimal y se almacena en la variable `hectareas`.
*   `dosis = float(input("Ingrese la dosis (L/ha): "))`: Pide al usuario que introduzca la dosis de fertilizante requerida por hectárea (en litros por hectárea). La entrada se convierte a flotante y se guarda en la variable `dosis`.

**Sección PROCESO:**
*   `fertilizante = hectareas * dosis`: Calcula la cantidad total de fertilizante necesaria. Multiplica las `hectareas` por la `dosis` por hectárea. El resultado (en litros de fertilizante) se almacena en la variable `fertilizante`.
*   `agua = fertilizante * 500`: Esta línea calcula la cantidad de agua necesaria, basándose en una relación de 1:500 (es decir, por cada litro de fertilizante, se necesitan 500 litros de agua). El resultado se guarda en la variable `agua`.

**Sección SALIDA:**
*   `print("Se necesitan", fertilizante, "L de fertilizante")`: Muestra la cantidad total de fertilizante requerida al usuario.
*   `print("Se necesitan", agua, "L de agua")`: Muestra la cantidad total de agua necesaria al usuario.

**Rendimiento de Pulpa en Frutas**

**Planteamiento del Problema:**

Una planta procesadora necesita saber el porcentaje de aprovechamiento. Se ingresa el peso total de la fruta comprada y el peso de los residuos (cáscaras y semillas).

**Requerimientos del problema:**

 1. Ingresar peso total(float).
 2. Ingresar residuos (float).
 3. Calcular pulpa:

 $$Pulpa=PesoTotal−Residuos$$
 4. Calcular rendimiento:

$$ R= \frac{Pulpa}{PesoTotal} \times 100$$


**Prueba del programa:**

Entrada:

*   500 kg total
*   150 kg residuos

Salida esperada:

*   350 kg pulpa
*   70% rendimiento







```python
# ENTRADA
peso_total = float(input("Ingrese el peso total de la fruta (kg): "))
residuos = float(input("Ingrese el peso de los residuos (kg): "))

# PROCESO
pulpa = peso_total - residuos
rendimiento = (pulpa / peso_total) * 100

# SALIDA
print("Pulpa obtenida:", pulpa, "kg")
print("Rendimiento:", rendimiento, "%")
```

### Explicación detallada del código para el Rendimiento de Pulpa en Frutas

**Sección ENTRADA:**
*   `peso_total = float(input("Ingrese el peso total de la fruta (kg): "))`: Solicita al usuario que ingrese el peso total de la fruta comprada en kilogramos. La entrada se convierte a un número decimal y se almacena en `peso_total`.
*   `residuos = float(input("Ingrese el peso de los residuos (kg): "))`: Pide al usuario que introduzca el peso de los residuos (cáscaras y semillas) en kilogramos. La entrada se convierte a flotante y se guarda en `residuos`.

**Sección PROCESO:**
*   `pulpa = peso_total - residuos`: Calcula el peso de la pulpa obtenida. Resta el peso de los `residuos` del `peso_total` de la fruta. El resultado se almacena en la variable `pulpa`.
*   `rendimiento = (pulpa / peso_total) * 100`: Esta línea calcula el rendimiento de la pulpa como un porcentaje. Divide el peso de la `pulpa` por el `peso_total` y luego multiplica el resultado por 100 para obtener el porcentaje. El resultado se guarda en la variable `rendimiento`.

**Sección SALIDA:**
*   `print("Pulpa obtenida:", pulpa, "kg")`: Muestra la cantidad de pulpa obtenida en kilogramos.
*   `print("Rendimiento:", rendimiento, "%")`: Muestra el porcentaje de rendimiento de la pulpa.
