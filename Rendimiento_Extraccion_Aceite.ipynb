**Control de Rendimiento en Extracción de Aceite**

Este ejercicio es ideal para introducir float, int y la concatenación de strings.

**Planteamiento:**

Una planta procesadora de palma de aceite necesita calcular el **Índice de Rendimiento de Extracción (IRE)**. El operario debe ingresar el peso total de los racimos de fruta fresca (RFF) procesados en kilogramos y la cantidad de aceite crudo obtenido en litros.

**Requerimientos**:

1. Solicitar el nombre del lote de producción (Tipo string).
2. Solicitar el peso de los racimos en kg (Tipo float). (pr)
3. Solicitar los litros de aceite obtenidos (Tipo float). (lt)
4. Calcular el rendimiento sabiendo que la densidad aproximada del aceite es de 0.92 kg/L. (r)
.
$$pa = lt * 0.92$$

$$r =(pa/pr)*100$$
**Prueba del programa:**

Si se ingresan 1000 kg de racimos y se obtienen 200 litros de aceite, el programa debe informar que el rendimiento del lote es del 18.4%.


El código para el 'Control de Rendimiento en Extracción de Aceite' se puede dividir en tres secciones principales: Entrada, Proceso y Salida. A continuación, se explica cada línea:

### Sección de Entrada:

*   `nombre=input("Ingrese el nombre del lote: ")`:
    *   Esta línea solicita al usuario que ingrese el nombre del lote de producción. El texto dentro de los paréntesis `("...")` es el mensaje que se muestra al usuario. La función `input()` lee lo que el usuario escribe desde la consola y lo almacena como una cadena de texto (string) en la variable `nombre`.

*   `while True:`:
    *   Inicia un bucle infinito que se ejecutará continuamente hasta que se encuentre una condición para salir (un `break`). Se utiliza para manejar errores de entrada.

*   `pr_input = input("Ingrese el peso de los racimos en kg (solo el número): ")`:
    *   Solicita al usuario el peso de los racimos en kilogramos. Similar a la entrada del nombre, `input()` captura lo que el usuario escribe, pero esta vez se espera un número. El mensaje aclara que solo se debe ingresar el número.

*   `try:`:
    *   Inicia un bloque `try-except`, que es una estructura para manejar errores. El código dentro del bloque `try` se intenta ejecutar. Si ocurre un error, la ejecución salta al bloque `except`.

*   `pr = float(pr_input)`:
    *   Intenta convertir la cadena de texto `pr_input` (el peso ingresado por el usuario) a un número de punto flotante (decimal) y lo asigna a la variable `pr`. Si `pr_input` no puede convertirse a un número (ej. contiene letras), se produce un `ValueError`.

*   `break`:
    *   Si la conversión a `float` es exitosa (es decir, no hubo `ValueError`), esta instrucción detiene el bucle `while True`, ya que se ha obtenido una entrada válida.

*   `except ValueError:`:
    *   Si se produce un `ValueError` durante la ejecución del bloque `try` (es decir, el usuario no ingresó un número válido), el programa salta a este bloque.

*   `print("Entrada inválida. Por favor, ingrese solo el valor numérico del peso.")`:
    *   Si ocurre un `ValueError`, esta línea imprime un mensaje de error informativo para el usuario, indicando que debe ingresar solo números. Luego, el bucle `while True` se repite, solicitando la entrada nuevamente.

*   Las líneas para `lt` (litros de aceite) siguen exactamente la misma lógica y estructura que las líneas para `pr` (peso de los racimos), asegurando que el usuario ingrese un valor numérico válido para los litros de aceite obtenidos.

### Sección de Proceso:

*   `pa = lt * 0.92`:
    *   Esta línea calcula el peso del aceite crudo (`pa`) en kilogramos. Se multiplica el volumen de aceite en litros (`lt`) por su densidad aproximada, que es 0.92 kg/L. El resultado se almacena en la variable `pa`.

*   `r = (pa/pr) * 100`:
    *   Esta línea calcula el Índice de Rendimiento de Extracción (`r`). Se divide el peso del aceite obtenido (`pa`) por el peso total de los racimos procesados (`pr`), y el resultado se multiplica por 100 para expresarlo como un porcentaje. El valor se guarda en la variable `r`.

### Sección de Salida:

*   `print("El rendimiento del lote", nombre, "es de", r, "%")`:
    *   Esta línea muestra el resultado final al usuario. La función `print()` concatena y muestra varias cadenas de texto y el valor de las variables `nombre` y `r` para formar un mensaje legible que informa el rendimiento del lote.

```python
# ENTRADA: Solicitar el nombre del lote de producción (Tipo string).
# Solicitar el peso de los racimos en kg (Tipo float). (pr)
# Solicitar los litros de aceite obtenidos (Tipo float). (lt)
nombre=input("Ingrese el nombre del lote: ")
pr=float(input("Ingrese el peso de los racimos en kg: "))
lt=float(input("Ingrese los litros de aceite obtenidos en litros: "))

# PROCESO: Calcular peso del aceite (pa) y el rendimiento (r)
pa = lt * 0.92
r = (pa/pr) * 100

# SALIDA: Informar el rendimiento del lote
print("El rendimiento del lote", nombre, "es de", r, "%")
```
